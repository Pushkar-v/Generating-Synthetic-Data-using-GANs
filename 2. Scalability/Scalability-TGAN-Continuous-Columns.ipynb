```python
import warnings
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
warnings.filterwarnings("ignore")


import time
import tensorflow as tf
import tensorflow.python.util.deprecation as deprecation
deprecation._PRINT_DEPRECATION_WARNINGS = False


import pandas as pd
import numpy as np
from tgan.model import TGANModel
```

```python
data = pd.read_csv('D:\Scalability Check\ContinuousData.csv', sep = ',')
data.shape
```

```python
data.head()
```

## Altering Number of Columns

### Function

```python
tf.reset_default_graph()
a1 = time.time()
tgan = TGANModel(continuous_columns = list(range(1,26)), restore_session=False)
tgan.fit(data.iloc[:,1:6])
b1 = time.time()
print('time : ', b1-a1)
print("Number of Rows : ", data.iloc[:,1:6].shape)   
```
