## Altering Number of Rows

### Function

```python
def timing(data, lst):
    # tf.reset_default_graph()
    a1 = time.time()
    tgan = TGANModel(continuous_columns = lst, restore_session=False)
    tgan.fit(data)
    b1 = time.time()
    print('time : ', b1-a1)
    print("Number of cols : ", data.shape[1])    
```
